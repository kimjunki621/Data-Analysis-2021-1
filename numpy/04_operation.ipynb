# Numpy operation

```python
import numpy as np
```

## axis

```python
matrix = np.arange(9).reshape(3,3)
matrix
```

```python
matrix.sum()
```

```python
matrix.sum(axis=0) # 행
```

```python
matrix.sum(axis=1) # 열
```

## mean & std (평균 및 표준편차)

```python
matrix.mean()
```

```python
matrix.mean(axis=0)
```

```python
# Standard deviation
matrix.std()
```

```python
matrix.std().round(4) # 반올림 소숫점 이하 4자리
```

## Mathematical functions

```python
matrix = np.arange(9).reshape(3,3)
np.exp(matrix)
```

## Concatenate(Array를 합치는 함수)

```python
a = np.array([[1,2],[3,4]])
a
```

```python
b = np.array([[5,6]])
b
```

```python
np.vstack((a,b))
```

```python
np.concatenate((a,b), axis=0)
```

```python
c = np.arange(6).reshape(2,3)
c
```

```python
np.hstack((a,c))
```

```python
np.concatenate((a,c),axis=1)
```

```python
np.hstack((a,b.T))
```

## dot product

```python
a = np.arange(6).reshape(3,2)
a
```

```python
b = np.arange(6).reshape(2,3)  # b = a.T
b
```

```python
a.dot(b)
```

```python
b.dot(a)
```

```python
np.dot(a,b)
```

```python
a * scalar
```
