# Pandas의 시각화 기능

```python
import numpy as np 
import pandas as pd 
```

## plot() method

```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot();
```

```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline

mpl.rcParams['axes.unicode_minus'] = False   # minus 표시
mpl.rcParams["font.family"] = 'Malgun Gothic'
```

```python
df1.plot()
plt.title("Pandas의 Plot메소드 사용 예")
plt.xlabel("시간")
plt.ylabel("Data")
plt.show()
```

## 다양한 플롯

```python
import seaborn as sns 
iris = sns.load_dataset("iris")  # 아이리스꽃 데이터
titanic = sns.load_dataset("titanic")  # 타이타닉호 데이터
```

```python
iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris.sepal_length[:20].plot.bar(rot=0)
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title("각 종의 Feature별 평균")
plt.xlabel("평균")
plt.ylabel("종")
plt.ylim(0, 8)
plt.show()
```

```python
df2.T
```

```python
df2.plot.bar(rot=0)
plt.title("각 Feature의 품종 별 평균")
plt.xlabel("평균")
plt.ylabel("종")
plt.ylim(0, 8)
plt.show()
```

```python
df3 = titanic['class'].value_counts()
df3
```

```python
df3.plot.pie(autopct='%.2f%%')
plt.title("선실별 승객 수 비율")
plt.axis('equal')
plt.show()
```

```python
iris.boxplot(by='species')
plt.tight_layout(pad=3, h_pad=1)  # pad 첫줄간격 / h_pad 간격
plt.show()
```
