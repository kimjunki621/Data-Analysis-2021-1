## 다양한 그래프

```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline

mpl.rcParams['axes.unicode_minus'] = False   # minus 표시
mpl.rcParams["font.family"] = 'Malgun Gothic'
```

### 1. 산점도(Scatter Plot)

```python
import numpy as np 
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y)
plt.show()
```

```python
colors =np.random.rand(100)
area = np.pi * (15 * np.random.rand(100)) ** 2
plt.scatter(x, y, label='Samples', s=area, c=colors, alpha=0.5)
plt.show()
```

### 2. Bar Chart

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

### 3. 히스토그램(Histogram)

```python
np.random.seed(0)
x = np.random.randn(1000)
plt.title("Histogram")
arrays, bins, patches = plt.hist(x, bins=10)
plt.show()
```

### 4. Box Plot

```python
x1 = np.random.randn(1000)
x2 = np.random.normal(loc=10, scale=2, size=1000)
plt.boxplot([x1, x2])
plt.show()
```

### 5. Pie Chart

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
